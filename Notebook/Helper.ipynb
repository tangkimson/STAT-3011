```python
import pandas as pd

# Set pandas options to display full content
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.max_colwidth', None)  # Show full content of each column

# Read the CSV file
file_path = '../data/messenger_reviews.csv'
data = pd.read_csv(file_path)

# Ensure the "at" column is treated as datetime
data['at'] = pd.to_datetime(data['at'], errors='coerce')

# Sort the data chronologically by the "at" column
data = data.sort_values(by='at')

# Specify the range of lines to print
start_line = 300  # Adjust as needed
end_line = 500  # Adjust as needed

# Display the specified range of rows as a table
data.iloc[start_line:end_line].style.set_table_attributes('style="width:100%"')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
file_path = '../data/cleaned_data/average_monthly_rating_messenger.csv'
data = pd.read_csv(file_path)

# Plot the data
plt.figure(figsize=(20, 6))
plt.plot(data['Month-Year'], data['averageRating'], marker='o', color='b', label='Average Rating')
plt.xlabel('Month-Year', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)
plt.title('Average Rating Over Time', fontsize=14)
plt.xticks(rotation=45, fontsize=10)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
